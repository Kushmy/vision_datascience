### Few more methods

#### apply() method with lambda keyword

```python
historic_df['Age_in_Days'] = historic_df.Age.apply(lambda x: x * 365.25) # adding a new column
```

```python
historic_df.head()
```

```python
historic_df.Age_in_Days = historic_df.Age_in_Days.apply(lambda x: int(x)) # converting existing column
```

```python
historic_df.head()
```

```python
historic_df.tail()
```

```python
historic_df_copy = historic_df.copy() 

# explicitly create a copy, anything we do now with historic_df_copy will not affect historic_df
```

```python
historic_df_copy.shape
```

#### replace() method

```python
historic_df_copy.Exam_Result.replace(1, 0)
```

```python
historic_df_copy.head() # data in dataframe did not change after replace()
```

```python
historic_df_copy.Exam_Result = historic_df_copy.Exam_Result.replace(1, 0) 

# this reassignment will cause the data to change
```

```python
historic_df_copy.head()
```

```python
historic_df_copy.Exam_Result.replace(0, 1) # once again, this will not change the original data
```

```python
historic_df_copy.head()
```

```python
historic_df_copy.Exam_Result.replace(0, 1, inplace=True) # this will change the original data
```

```python
historic_df_copy.head()
```

#### duplicated() method 

```python
historic_df_copy.duplicated().sum() # we want to count the number of duplicate rows, so using sum()
```

```python
historic_df_copy.tail()
```

```python
historic_df_copy.ix[1000] = ['S1', 41, 1, 1, 3, 23, 0, 1, 41, 1, 14975] 

# adding a new row, this is NOT the preferred way, mostly because of the warning below!
```

```python
historic_df_copy.loc[1001] = ('S1', 41, 1, 1, 3, 23, 0, 1, 41, 1, 14975)

# this is better, also does not give any warnings!
# but remember that this is good only if you need to add a row or two and rarely
# if you need to add many rows, there are other better ways
```

```python
historic_df_copy.tail()
```

```python
historic_df_copy.duplicated().sum() # the 2 below means there is 1 original and 2 duplicates
```

#### drop_duplicates() method

```python
historic_df_copy.drop_duplicates() 

# by default, it will keep the first occurance (original) and remove the other occurances (duplicates)
```

```python
historic_df_copy.head() 
```

```python
historic_df_copy.tail() # once again, we see the the duplicate is not removed!
```

```python
historic_df_copy.drop_duplicates(inplace=True) # this will actually remove the duplicate due to inplace=True
```

```python
historic_df_copy.tail()
```

#### The following shows that the dataframe we started with - historic_df - has not been affected by the operations we performed on historic_df_copy
