#### Converting string type to category using the astype() method, then converting to numbers using .cat.codes property

```python
raw_data['location_codes'] = raw_data.location.astype('category').cat.codes 

# adding as a new column, not modifying original as getting back the original strings is a more complex task (if you need them later)
```

#### Summary statistics using the describe() method:
<br>
##### &emsp; Count: The number of items
##### &emsp; Mean: The arithmetic mean, also knon simply as 'average'
##### &emsp; Std: The standard deviation - a unit to measure how far a data point is from the mean.
##### &emsp; Min: The minimum value.
##### &emsp; 25%: Also know as the 1st quartile or 25th empirical quartile, it is a number that indicates that 25% of the values are below this number.
##### &emsp; 50%: Also know as the 2nd quartile or 50th empirical quartile, it is a number that indicates that 50% of the values are below this number.
##### &emsp; 75%: Also know as the 3nd quartile or 75th empirical quartile, it is a number that indicates that 75% of the values are below this number.
##### &emsp; Max: The maximum value.

```python
raw_data.describe()
```

##### Create a new column to indicate the number of seconds after offset

```python
svrtime_min = raw_data.svrtime.min() # this is our offset timestamp
```

```python
raw_data['time_diff_from_offset_in_seconds'] = raw_data.svrtime.apply(lambda x: (x - svrtime_min).total_seconds()) 

# calculating time difference from offset (in seconds)
# this took only a couple of seconds!
```

```python
raw_data.head()
```

#### groupby() method: useful to find unique combinations, also to fetch data in groups

```python
raw_data_gb_latlon = raw_data.groupby(['latitude', 'longitude'])
```

```python
raw_data_gb_latlon.ngroups # property to show number of groups, in this case it thhe number of unique locations
```
